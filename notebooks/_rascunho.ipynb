```python
def count_hobbies(hobbies, word):
    counter = 0
    for hobby in hobbies:
        if word in hobby.split():
            counter += 1
    return counter

count_hobbies(["Tennis", "Tables Tennis"], "tennis")
```

```python
from pprint import pprint

class MapMatrix:
    def __init__(self, map):
        self.matrix = map

    def get_string_map(self, bool_map):
        string_map = [[ "R" if col else "E" for col in row] for row in bool_map]
        return string_map

    def draw_string_map(self, string_map):
        map_dict = {
            "R": "[]",
            "E": "  ",
            "P": "██",
        }
        
        print("   " + " ".join(
            [str(x) if x > 9 else " " + str(x) for x in list(range(0, len(string_map[0])))]
        ))

        for i, row in enumerate(string_map):
            row_head = str(i) if i > 9 else " " + str(i)
            print(row_head + " " + " ".join([map_dict[j] for j in string_map[i]]))            
        print('\n')    

    def draw_map(self):
        m = self.get_string_map(self.matrix)
        self.draw_string_map(m)

    def draw_path(self, path):
        m = self.get_string_map(self.matrix)
        for step in path:
            m[step[0]][step[1]] = "P"
        self.draw_string_map(m)
        
        # print("   " + " ".join(
        #     [str(x) if x > 9 else " " + str(x) for x in list(range(0, len(m[0])))]
        # ))
        
        # for i, row in enumerate(m):
        #     row_head = str(i) if i > 9 else " " + str(i)
        #     print(row_head + " " + " ".join(["██" if j else "  " for j in row]))
        # print('\n')

    def get_connected_roads(self, row, column):
        roads = []
        
        try:
            down = self.matrix[row + 1][column]
            if down:
                roads.append((row + 1, column))
        except IndexError:
            down = False
        
        try:
            right = self.matrix[row][column + 1]
            if right:
                roads.append((row, column + 1))
        except IndexError:
            right = False
    
       
    
        if column > 0:
            try:
                left = self.matrix[row][column - 1]
                if left:
                    roads.append((row, column - 1))
            except IndexError:
                left = False

        if row > 0:
            try:
                up = self.matrix[row - 1][column]
                if up:
                    roads.append((row - 1, column))
            except IndexError:
                up = False

        return roads

    
    def route_exists(self, from_row, from_column, to_row, to_column, path=None, dead_ends=None):
        print(f'MapMatrix.route_exists(from=({from_row},{from_column}), to=({to_row},{to_column}), {path=}, {dead_ends=})')
        if not path:
            path = []
        if not dead_ends:
            dead_ends = []
        if (from_row, from_column) in path:
            return False
        path.append((from_row, from_column))

        print(f'    {path=}')
        print(f'    {dead_ends=}')
        
        roads = self.get_connected_roads(from_row, from_column)
        print(f'    {roads=}')

        self.draw_path(path)
        
        if (to_row, to_column) in roads:
            path.append((to_row, to_column))
            print('Route: ', path)
            return True
        
        for road in roads:
            if road not in dead_ends:
                if road not in path:
                    if self.route_exists(road[0], road[1], to_row, to_column, path[:], dead_ends):
                        return True
                    else:
                        print(f'    dead_end found: {road=}')
                        dead_ends.append(road)
                    
        # dead_ends.append((from_row, from_column))
        # print(f'    dead_end found: ({from_row},{from_column})')
        return False


def route_exists(from_row, from_column, to_row, to_column, map_matrix):
    print(f'\nroute_exists(from=({from_row},{from_column}), to=({to_row},{to_column}))')
    m = MapMatrix(map_matrix)
    return m.route_exists(from_row, from_column, to_row, to_column)


map_matrix1 = [
    [True, False, False],
    [True, True, False],
    [False, True, True]
];
map_matrix2 = [
    [True, True, True, False, False, False],
    [True, True, True, True, True, True],
    [False, False, True, True, True, True]
];
map_matrix3 = [
    map_matrix2[0] + map_matrix2[1] + map_matrix2[2],
    map_matrix2[1] + map_matrix2[2] + map_matrix2[0],
    map_matrix2[2] + map_matrix2[0] + map_matrix2[1],
    map_matrix2[1] + map_matrix2[2] + map_matrix2[0],
    map_matrix2[0] + map_matrix2[1] + map_matrix2[2],
]
map_matrix4 = [
    map_matrix3[0] + map_matrix3[1], map_matrix3[1] + map_matrix3[2], map_matrix3[2] + map_matrix3[0], map_matrix3[1] + map_matrix3[2],
    map_matrix3[0] + map_matrix3[1], map_matrix3[0] + map_matrix3[1], map_matrix3[1] + map_matrix3[2], map_matrix3[2] + map_matrix3[0],
    map_matrix3[1] + map_matrix3[2], map_matrix3[0] + map_matrix3[1], map_matrix3[0] + map_matrix3[1], map_matrix3[1] + map_matrix3[2],
    map_matrix3[2] + map_matrix3[0], map_matrix3[1] + map_matrix3[2], map_matrix3[0] + map_matrix3[1], map_matrix3[0] + map_matrix3[1],
    map_matrix3[1] + map_matrix3[2], map_matrix3[2] + map_matrix3[0], map_matrix3[1] + map_matrix3[2], map_matrix3[0] + map_matrix3[1],
]

m3 = MapMatrix(map_matrix3)
# m3.draw_map()
# path=[(0, 0), (0, 1), (0, 2), (1, 2), (1, 3), (1, 4), (1, 5), (2, 5), (2, 6), (2, 7), (2, 8), (3, 8), (3, 9), (3, 10), (3, 11)]
# m3.draw_path(path)
m3.route_exists(0, 0, 4, 17)

m4 = MapMatrix(map_matrix4)
# m4.draw_map()

# path=[(0, 0), (0, 1), (0, 2), (1, 2), (1, 3), (1, 4), (1, 5), (2, 5), (2, 6), (2, 7), (2, 8), (3, 8), (3, 9), (3, 10), (3, 11)]
# m4.draw_path(path)
```
