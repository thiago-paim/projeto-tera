```python
# Conta quantos tweets existem da ErikakHilton na base
query = """
    SELECT COUNT(t.id)
    FROM tweets_tweet AS t
    JOIN tweets_twitteruser AS u
        ON t.user_id = u.id
    WHERE u.username == 'ErikakHilton'
    ;
"""
pd.read_sql_query(query, conn)
```

```python
# Seleciona tweets por username
query = """
    SELECT t.published_at, t.content, u.username
    FROM tweets_tweet AS t
    JOIN tweets_twitteruser AS u
        ON t.user_id = u.id
    WHERE u.username == 'ErikakHilton'
    ;
"""
pd.read_sql_query(query, conn)
```

```python
# Seleciona 10 primeiros tweets que responderam a um username
query = """
    SELECT 
        reply.published_at, 
        reply.content, 
        reply_user.username, 
        tweet.published_at, 
        tweet.content, 
        tweet_user.username
    FROM tweets_tweet AS reply
    JOIN tweets_tweet AS tweet
        ON reply.in_reply_to_tweet_id = tweet.id
    JOIN tweets_twitteruser AS reply_user
        ON reply.user_id = reply_user.id
    JOIN tweets_twitteruser AS tweet_user
        ON tweet.user_id = tweet_user.id
    WHERE
        tweet_user.username = 'ErikakHilton'
    LIMIT 10
    ;
"""
df = pd.read_sql_query(query, conn)
df.head()
```

```python
# Mesma consulta acima, mas renomeando as colunas para evitar nomes repetidos
query = """
    SELECT 
        reply.published_at AS reply_published_at, 
        reply.content AS reply_content, 
        reply_user.username AS reply_username, 
        tweet.published_at AS tweet_published_at, 
        tweet.content AS tweet_content, 
        tweet_user.username AS tweet_username
    FROM tweets_tweet AS reply
    JOIN tweets_tweet AS tweet
        ON reply.in_reply_to_tweet_id = tweet.id
    JOIN tweets_twitteruser AS reply_user
        ON reply.user_id = reply_user.id
    JOIN tweets_twitteruser AS tweet_user
        ON tweet.user_id = tweet_user.id
    WHERE
        tweet_username = 'ErikakHilton'
    LIMIT 10
    ;
"""
df = pd.read_sql_query(query, conn)
df.head()
```

```python
# Contando quantas linhas a query retorna antes de pegar todos os resultados, para checar se n√£o sobrecarregar a mem√≥ria
query = """
    SELECT 
        count(reply.id)
    FROM tweets_tweet AS reply
    JOIN tweets_tweet AS tweet
        ON reply.in_reply_to_tweet_id = tweet.id
    JOIN tweets_twitteruser AS reply_user
        ON reply.user_id = reply_user.id
    JOIN tweets_twitteruser AS tweet_user
        ON tweet.user_id = tweet_user.id
    WHERE
        tweet_user.username = 'ErikakHilton'
    ;
"""
df = pd.read_sql_query(query, conn)
df.head()
```

```python
# Agora sim, buscando todas os tweets que responderam a um usuario
query = """
    SELECT 
        reply.published_at AS reply_published_at, 
        reply.content AS reply_content, 
        reply_user.username AS reply_username, 
        tweet.published_at AS tweet_published_at, 
        tweet.content AS tweet_content, 
        tweet_user.username AS tweet_username
    FROM tweets_tweet AS reply
    JOIN tweets_tweet AS tweet
        ON reply.in_reply_to_tweet_id = tweet.id
    JOIN tweets_twitteruser AS reply_user
        ON reply.user_id = reply_user.id
    JOIN tweets_twitteruser AS tweet_user
        ON tweet.user_id = tweet_user.id
    WHERE
        tweet_username = 'ErikakHilton'
    ;
"""
df = pd.read_sql_query(query, conn)
df.info()
```
